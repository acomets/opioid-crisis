```python
import folium
import folium.plugins
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/overdoses.csv', thousands=',')
```

```python
df['Deaths per Million'] = df['Deaths'] / df['Population'] * 1E6
```

```python
m = folium.Map(location=[42.0285, -93.65], zoom_start=4)
m.choropleth(
    geo_data='../data/us-states.json',
    data=df,
    columns=['State', 'Deaths per Million'],
    key_on='feature.properties.name',
    threshold_scale=list(np.linspace(0, df['Deaths per Million'].max(), num=6)),
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Overdoses (Deaths per Million)',
    highlight=True
)
```
