```python
import folium
import folium.plugins
import pandas as pd
```

```python
df = pd.read_csv('overdoses.csv', thousands=',')
```

```python
df['Deaths per Million'] = df['Deaths'] / df['Population'] * 1E6
```

```python
m = folium.Map(location=[42.0285, -93.65], zoom_start=4)
m.choropleth(
    geo_data='us-states.json',
    name='choropleth',
    data=df,
    columns=['State', 'Deaths per Million'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Deaths per Million'
)
```
